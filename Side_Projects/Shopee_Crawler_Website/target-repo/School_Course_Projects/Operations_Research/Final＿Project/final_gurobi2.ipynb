```python
import numpy as np
import pandas as pd
# Read Excel file and load data
file_path = "Instances/Group2/Instance2_0_real.xlsx"

data = import_data.load_file(file_path)

budget = data['BUDGET']
num_days = data['D']
num_hotels = data['M']  
num_attractions = data['I']
num_places = data['P']
num_transport_modes = data['K']

# S_M = [m for m in range(1, data['M'] + 1)]  #  the first hotel is the starting airport and the last hotel is the leaving airport
# S_I = [i for i in range(1, data['I'] + 1)]
# S_P = [p for p in range(1, data['I'] + data['M'] + 1)]
# S_D = [d for d in range(1, data['D'] + 1)]
# S_K = [k for k in range(1, data['K'] + 1)]

S_M = [m for m in range(0, data['M'])]  #  the first hotel is the starting airport and the last hotel is the leaving airport
S_I = [i for i in range(0, data['I'])]
S_P = [p for p in range(0, data['I'] + data['M'])]
S_D = [d for d in range(0, data['D'])]
S_K = [k for k in range(0, data['K'])]


# 第一天抵達機場時間
arrival_time_day1 = data['Day1Arrival'] # 需要從 Excel 讀取

# 每天離開旅館的時間 (不包含第一天和最後一天)
departure_time_hotel = data['LH'] # 需要從 Excel 讀取
departure_time_hotel = departure_time_hotel.reset_index(drop=True)

# 每天返回旅館的最晚時間
return_time_hotel = data['RH']    # 需要從 Excel 讀取
return_time_hotel = return_time_hotel.reset_index(drop=True) 

# 每個景點的停留時間 (表格，索引為景點 i)
stay_duration = data['STAY']      # 需要從 Excel 讀取 (DataFrame 或 dict)
stay_duration = stay_duration.reset_index(drop=True)

# 每個景點的成本 (表格，索引為景點 i)
spot_costs = data['C_ATTRACTION']         # 需要從 Excel 讀取 (DataFrame 或 dict)

# 每個景點的幸福值 (表格，索引為景點 i)
happiness_values = data['HAPPINESS']    # 需要從 Excel 讀取 (DataFrame 或 dict)
happiness_values = happiness_values.reset_index(drop=True)

# 每個景點每天的開放時間 (表格，索引為景點 i 和天 d)
open_times = data['ATTRACTION_OPEN']          # 需要從 Excel 讀取 (DataFrame 或 dict)

# 每個景點每天的關閉時間 (表格，索引為景點 i 和天 d)
close_times = data['ATTRACTION_CLOSE']         # 需要從 Excel 讀取 (DataFrame 或 dict)

# 旅館每天的住宿成本 (表格，索引為旅館 m 和天 d，不包含起始和結束機場)
hotel_costs = data['C_HOTEL']        # 需要從 Excel 讀取 (DataFrame 或 dict)

# 地點 p 到地點 q 使用交通方式 k 的交通時間 (表格，索引為 p, q, k)
transport_times = data['DURATION']      # 需要從 Excel 讀取 (DataFrame 或 MultiIndex DataFrame 或 dict)
transport_times = np.stack(list(transport_times.values()))
# 地點 p 到地點 q 使用交通方式 k 的交通成本 (表格，索引為 p, q, k)
transport_costs = data['C_TRAVEL']     # 需要從 Excel 讀取 (DataFrame 或 MultiIndex DataFrame 或 dict)
transport_costs = np.stack(list(transport_costs.values()))

#重要!補0
stay_duration = pd.concat([stay_duration, pd.Series(0 for _ in range(num_hotels))], ignore_index=True)

print(f"Budget: {budget}")
print(f"Number of days: {num_days}")
print(f"Number of hotels: {num_hotels}")
print(f"Number of attractions: {num_attractions}")
print(f"Number of places: {num_places}")
print(f"Number of transport modes: {num_transport_modes}")

print("Sets:")
print(f"S_M: {S_M}")
print(f"S_I: {S_I}")
print(f"S_P: {S_P}")
print(f"S_D: {S_D}")
print(f"S_K: {S_K}")

M = S_M
I = S_I
P = S_P
D = S_D
K = S_K


print(f"Day 1 Arrival Time: {arrival_time_day1}")
print(f"Departure Time from Hotel: {departure_time_hotel}")
print(f"Return Time to Hotel: {return_time_hotel}")
print(f"Stay Duration: {stay_duration}")
print(f"Spot Costs: {spot_costs}")
print(f"Happiness Values: {happiness_values}")
print(f"Open Times: {open_times}")
print(f"Close Times: {close_times}")
print(f"Hotel Costs: {hotel_costs}")
print(f"Transport Times: {transport_times}")
print(f"Transport Costs: {transport_costs}")
```

```python
bigM = 1440
# ----- 3. 建立 Gurobi 模型 -----
import gurobipy as gp
from gurobipy import GRB
print("Creating model...")
model = gp.Model("travel_planning")
print("Model created.")

# ----- 4. 定義決策變數 -----
x = model.addVars(S_I, S_D, vtype=GRB.BINARY, name="visit")
t = model.addVars(S_P, S_P, S_K, S_D, vtype=GRB.BINARY, name="travel")
ha = model.addVars(S_M, S_D, vtype=GRB.BINARY, name="arrive_hotel")
hl = model.addVars(S_M, S_D, vtype=GRB.BINARY, name="leave_hotel")
a = model.addVars(S_P, S_D, vtype=GRB.CONTINUOUS, name="arrival_time")
lt = model.addVars(S_D, vtype=GRB.CONTINUOUS, name="arrival_time")
print("Defined variables:")

# ----- 5. 設定目標函數 -----
# i 是從 num_hotels + 2 機場 + 1 開始算起，例如 instance0 中 i 從 4 開始
# 但是 happiness_values 等參數的 index 是從 0 開始算起
# 所以在 index 這邊就扣會回去
model.setObjective(gp.quicksum(happiness_values[i] * x[i, d] for d in S_D for i in S_I), GRB.MAXIMIZE)
print("Objective function set.")   


# ----- 6. 加入約束條件 -----
#預算約束 (equation 2)
model.addConstr(
    gp.quicksum(transport_costs[k,p,q] * t[p, q, k, d] for d in S_D for p in S_P for q in S_P if p != q for k in S_K)
    + gp.quicksum(hotel_costs[d,m] * ha[m, d] for m in S_M[1:-1] for d in S_D) 
    + gp.quicksum(spot_costs[i] * x[i, d] for d in S_D for i in S_I) <= budget,
    "budget_constraint"
    )
# 第一天抵達機場時間 (equation 3) delete
#model.addConstr(a[0, 0] == arrival_time_day1, "arrival_day1")

# 其他天離開旅館時間 (equation 4) delete
# for d in D[1:]:
#     for m in M[1:-1]:
#         model.addConstr(a[m, d] == departure_time_hotel[d] * hl[m, d], f"departure_hotel_{m}_{d}")

# 抵達景點時間下限 (equation 5)
for i in I:
    for d in D:
        model.addConstr(a[i, d] >= departure_time_hotel[d] * x[i, d], f"arrival_spot_lower_{i}_{d}")

# 景點開放時間約束 (equation 6)
for i in I:
    for d in D:
        model.addConstr(a[i, d] >= open_times[i,d] * x[i, d], f"open_time_{i}_{d}")

# 定義一個足夠大的時間值，用於取代無限大 (Infinity)
# 假設您的時間單位是小時，一天有 24 小時，最多5天，那麼 1000 是一個保守的足夠大的數
MAX_POSSIBLE_TIME = 10000.0 # 例如，用 10000 分鐘代表"無限大"

#景點關閉時間約束 (equation 7)
for i in I:
    for d in D:
        # 將 float('inf') 替換為 MAX_POSSIBLE_TIME
        model.addConstr(a[i, d] + stay_duration[i] * x[i, d] <= close_times[i,d] * x[i, d], f"close_time_{i}_{d}")

# --- 新增的住宿地點相關約束 ---

# 不可以跑回去睡機場 (離開)
# h^l_{1d} = 0, forall d in D \setminus \{1\}
for d in D:
    if d != 0:
        model.addConstr(hl[0, d] == 0, f"no_leave_start_airport_{d}")

# 不可以跑回去睡機場 (抵達)
# h^a_{n_md} = 0, forall d in D \setminus \{n_d\}
for d in D:
    if d != num_days - 1: # num_days 是您之前定義的總天數
        model.addConstr(ha[num_hotels - 1, d] == 0, f"no_arrive_end_airport_{d}")


# ... (您後續的其他約束，例如每天只能住一間旅館等)

# 第一天返回旅館時間約束 (equation 8)
model.addConstr(
    arrival_time_day1 +
    gp.quicksum(stay_duration[i] * x[i, 0] for i in I) +
    gp.quicksum(transport_times[k,p,q] * t[p, q, k, 0] for p in P for q in P if p != q for k in K) <= return_time_hotel[0],
    "return_day1"
)

# 其他天返回旅館時間約束 (equation 9)
for d in D[1:]:
    model.addConstr(
        lt[d] + 
        gp.quicksum(stay_duration[i] * x[i, d] for i in I) +
        gp.quicksum(transport_times[k,p,q] * t[p, q, k, d] for p in P for q in P if p != q for k in K) <= return_time_hotel[d],
        f"return_day_{d}"
    )

# 返回旅館時間約束-2 (new constraint 9-2)
for d in D:
    for m in M:
        model.addConstr(
            a[num_attractions+m,d] <= return_time_hotel[d]+ bigM * (1 - ha[m,d]),
            f"return_to_hotel_{m}_{d}"
        )


# 同一天內地點間移動的時間約束 (equation 10)

#從hotel 出發 (new equation)
for d in D:
    for q in P:
        travel_time_expr = gp.quicksum(transport_times[k,num_attractions+m,q] * t[num_attractions+m, q, k, d] for k in K for m in M)
        lhs = lt[d]  + travel_time_expr - a[q, d]
        rhs = bigM * (1 - gp.quicksum(t[num_attractions+m, q, k, d] for k in K for m in M))
        #print(f"p=hotel, q={q}, d={d}, lhs={lhs}, rhs={rhs}") # <--- 在這裡添加 print 語句
        model.addConstr(lhs <= rhs, f"travel_time_hotel_{q}_{d}")

#from tourist spots to other places (updated equation 10)
for i in I:
    for q in P:
        if i != q:
            for d in D:
                travel_time_expr = gp.quicksum(transport_times[k,i,q] * t[i, q, k, d] for k in K)
                lhs = a[i, d] + stay_duration[i] + travel_time_expr - a[q, d]
                rhs = bigM * (1 - gp.quicksum(t[i, q, k, d] for k in K))
                #print(f"p={i}, q={q}, d={d}, lhs={lhs}, rhs={rhs}") # <--- 在這裡添加 print 語句
                model.addConstr(lhs <= rhs, f"travel_time_{i}_{q}_{d}")


# 每個景點最多參觀一次 (equation 11)
for i in I:
    model.addConstr(gp.quicksum(x[i, d] for d in D) <= 1, f"visit_once_{i}")

# 每天必須抵達一個旅館(equation 12)
for d in D:
    model.addConstr(gp.quicksum(ha[m, d] for m in M) == 1, f"arrive_one_hotel_{d}")

# 如果今晚抵達旅館 m，明天早上必須從同一旅館 m 出發(最後一天除外) (equation 13)
for m in M:
    for d in D[:-1]:
        model.addConstr(ha[m, d] == hl[m, d + 1], f"hotel_continuity_{m}_{d}")

# 第一天早上必須離開機場 (equation 14)
model.addConstr(hl[0, 0] == 1, "leave_airport_day1")

# 最後一天晚上必須抵達機場 (equation 15)
model.addConstr(ha[num_hotels - 1, num_days - 1] == 1, "arrive_airport_last_day")

# 離開景點後的去向 (equation 16) - 修改為逐日約束
for i in I:
    for d in D: # 新增 d 的外層循環
        model.addConstr(
            gp.quicksum(t[i, j, k, d] for j in I for k in K if i != j) + 
            gp.quicksum(t[i, num_attractions + m, k, d] for m in M for k in K) == x[i, d], 
            f"leave_spot_{i}_{d}" # 修改約束名稱以包含 d
        )

# 前往景點的起點 (equation 17) - 修改為逐日約束
for i in I:
    for d in D: # 新增 d 的外層循環
        model.addConstr(
            gp.quicksum(t[j, i, k, d] for j in I for k in K if i != j) + # 移除 d 的循環
            gp.quicksum(t[num_attractions + m, i, k, d] for m in M for k in K) == x[i, d], # 移除 d 的循環
            f"arrive_spot_{i}_{d}" # 修改約束名稱以包含 d
        )

# 早上離開旅館後的去向 (equation 18) -update:可以留在旅館裡
for m in M:
    for d in D:
        model.addConstr(
            gp.quicksum(t[num_attractions + m, q, k, d] for q in P for k in K) == hl[m, d],
            f"leave_hotel_to_spot_{m}_{d}"
        )

# 晚上離開景點(旅館)後要去旅館 (equation 19)
for m in M:
    for d in D:
        model.addConstr(
            gp.quicksum(t[p, num_attractions + m, k, d] for p in P for k in K) == ha[m, d],
            f"arrive_hotel_from_spot_{m}_{d}"
        )

# 每個地方一天只能到達一次 (new equation)
for d in D:
    for q in P:
        model.addConstr(
            gp.quicksum(t[p, q, k, d] for p in P for k in K) <= 1,
            f"arrive_maximumn_once_at_each_spot{q}_on_each_day_{d}"
        )

# 每個地方只能用一種交通方式抵達 (new equation)
for d in D:
    for p in P:
        for q in P:
            model.addConstr(
                gp.quicksum(t[p, q, k, d] for k in K) <= 1,
                f"one transportation type{p}_{q}_{d}"
            )

#離開時間要大於使用者設定的時間 (new equation)
for d in D:
    model.addConstr(
        lt[d] >= departure_time_hotel[d],
        f"leavetime_{d}"
    )


# ----- 7. 求解模型 -----
model.optimize()
```

```python
# ----- 8. 輸出結果 -----
if model.status == GRB.OPTIMAL:
    print("\nOptimal solution found:") # 這是新增的
    print("--- Arrival Times (a[p, d]) ---") # 這是新增的
    for p in P: # 這是新增的
        for d in D: # 這是新增的
            if a[p,d].X is not None: # 這是新增的
                print(f"Arrival time at location {p} on day {d}: {a[p,d].X:.2f}") # 這是新增的
    print("---------------------------------") # 這是新增的

    # 您也可以選擇性地打印其他變數的值，例如 x[i,d] (表示景點是否被訪問)
    print("\n--- Visited Spots (x[i, d]) ---") # 這是新增的
    for i in I: # 這是新增的
        for d in D: # 這是新增的
            if x[i,d].X > 0.5: # 這是新增的
                print(f"Spot {i} visited on day {d}") # 這是新增的
    print("-------------------------------") # 這是新增的

    # 如果您還想看看交通方式 t 的選擇
    print("\n--- Chosen Travel Modes (t[p, q, k, d]) ---") # 這是新增的
    for p in P: # 這是新增的
        for q in P: # 這是新增的
            if p != q: # 這是新增的
                for d in D: # 這是新增的
                    for k in K: # 這是新增的
                        if t[p, q, k, d].X > 0.5: # 這是新增的
                            print(f"Travel from {p} to {q} using mode {k+1 if K else 1} on day {d}") # 這是新增的
    print("---------------------------------------------") # 這是新增的
    print("\n最佳化完成！")
    print(f"最大總幸福值：{model.objVal}")

    
# 輸出哪些景點在哪天參觀
    for d in D:
        print(f"\n第 {d} 天：")
        for i in I:
            if x[i, d].x > 0.5:
                print(f"  參觀景點 {i}")

    # 輸出每天的行程 (包含移動)
    for d in D:
        print(f"\n第 {d} 天詳細行程：")
        current_location = num_attractions if d == 0 else None # 第一天從機場開始
        leave_hotel = num_attractions if d == 0 else None
        if d >= 1:
            for m in M:
                if hl[m, d].x > 0.5:
                    leave_hotel = num_attractions+m
                    current_location = num_attractions+m
                    print(f"離開旅館 {m} (時間: {lt[d].x} 限制最早為 {departure_time_hotel[d]})")
                    break

        while True:
            moved = False
            for p in P:
                if p == current_location:
                    for q in P:
                        if p != q:
                            for k in K:
                                if t[p, q, k, d].x > 0.5:
                                    print(f"  從地點 {p} 移動到地點 {q} (交通方式 {k+1 if K else 1}, 花費時間: {transport_times[k][p][q]})")
                                    current_location = q
                                    if q in I and x[q, d].x > 0.5:
                                        print(f"  參觀景點 {q} (抵達時間: {a[q, d].x}, 停留時間: {stay_duration[q]})")
                                    moved = True
                                    break
                            if moved:
                                break
                    break
            if not moved or current_location == leave_hotel:
                # 檢查是否抵達旅館
                for m in M:
                    if ha[m, d].x > 0.5:
                        print(f" 抵達地點 {num_attractions+m} 也就是旅館 {m} (時間: {a[num_attractions+m, d].x},限制為RH={return_time_hotel[d]})")
                        current_location = num_attractions + m
                        moved = True
                        break
                if moved:
                    break # 當天沒有更多移動

elif model.status == GRB.INFEASIBLE:
    print("Model is infeasible (無可行解)")
    print("Consider checking constraints for conflicts.")
    model.computeIIS()  # <--- 取消註解這一行
    model.write("infeasible.ilp") # <--- 取消註解這一行
    print("Infeasible constraints written to infeasible.ilp")
    # 如果您還想將模型保存為LP格式以供其他工具查看 (可選)
    # model.write("full_model.lp")

else:
    print("模型沒有找到最佳解")
```

```python
def LR(
    S_I, S_D, S_P, S_K, S_M,
    I, D, P, K, M,
    num_attractions, num_hotels, num_days,
    happiness_values, transport_costs, hotel_costs, spot_costs, budget,
    departure_time_hotel, arrival_time_day1, stay_duration,
    open_times, close_times, return_time_hotel, transport_times
):
    bigM = 1440
    # ----- 3. 建立 Gurobi 模型 -----
    import gurobipy as gp
    from gurobipy import GRB
    print("Creating model...")
    model = gp.Model("travel_planning")
    print("Model created.")
    
    # ----- 4. 定義決策變數 Linear Relaxation Version -----
    x = model.addVars(S_I, S_D, vtype=GRB.CONTINUOUS, lb=0.0, ub=1.0, name="visit")
    t = model.addVars(S_P, S_P, S_K, S_D, vtype=GRB.CONTINUOUS, lb=0.0, ub=1.0, name="travel")
    ha = model.addVars(S_M, S_D, vtype=GRB.CONTINUOUS, lb=0.0, ub=1.0, name="arrive_hotel")
    hl = model.addVars(S_M, S_D, vtype=GRB.CONTINUOUS, lb=0.0, ub=1.0, name="leave_hotel")
    a = model.addVars(S_P, S_D, vtype=GRB.CONTINUOUS, lb=0.0, name="arrival_time")
    lt = model.addVars(S_D, vtype=GRB.CONTINUOUS, name="arrival_time")
    print("Defined variables:")
    
    # ----- 5. 設定目標函數 -----
    # i 是從 num_hotels + 2 機場 + 1 開始算起，例如 instance0 中 i 從 4 開始
    # 但是 happiness_values 等參數的 index 是從 0 開始算起
    # 所以在 index 這邊就扣會回去
    model.setObjective(gp.quicksum(happiness_values[i] * x[i, d] for d in S_D for i in S_I), GRB.MAXIMIZE)
    print("Objective function set.")   
    
    
    # ----- 6. 加入約束條件 -----
    #預算約束 (equation 2)
    model.addConstr(
        gp.quicksum(transport_costs[k,p,q] * t[p, q, k, d] for d in S_D for p in S_P for q in S_P if p != q for k in S_K)
        + gp.quicksum(hotel_costs[d,m] * ha[m, d] for m in S_M[1:-1] for d in S_D) 
        + gp.quicksum(spot_costs[i] * x[i, d] for d in S_D for i in S_I) <= budget,
        "budget_constraint"
        )
    # 第一天抵達機場時間 (equation 3) delete
    #model.addConstr(a[0, 0] == arrival_time_day1, "arrival_day1")
    
    # 其他天離開旅館時間 (equation 4) delete
    # for d in D[1:]:
    #     for m in M[1:-1]:
    #         model.addConstr(a[m, d] == departure_time_hotel[d] * hl[m, d], f"departure_hotel_{m}_{d}")
    
    # 抵達景點時間下限 (equation 5)
    for i in I:
        for d in D:
            model.addConstr(a[i, d] >= departure_time_hotel[d] * x[i, d], f"arrival_spot_lower_{i}_{d}")
    
    # 景點開放時間約束 (equation 6)
    for i in I:
        for d in D:
            model.addConstr(a[i, d] >= open_times[i,d] * x[i, d], f"open_time_{i}_{d}")
    
    # 定義一個足夠大的時間值，用於取代無限大 (Infinity)
    # 假設您的時間單位是小時，一天有 24 小時，最多5天，那麼 1000 是一個保守的足夠大的數
    MAX_POSSIBLE_TIME = 10000.0 # 例如，用 10000 分鐘代表"無限大"
    
    #景點關閉時間約束 (equation 7)
    for i in I:
        for d in D:
            # 將 float('inf') 替換為 MAX_POSSIBLE_TIME
            model.addConstr(a[i, d] + stay_duration[i] * x[i, d] <= close_times[i,d] * x[i, d], f"close_time_{i}_{d}")
    
    # --- 新增的住宿地點相關約束 ---
    
    # 不可以跑回去睡機場 (離開)
    # h^l_{1d} = 0, forall d in D \setminus \{1\}
    for d in D:
        if d != 0:
            model.addConstr(hl[0, d] == 0, f"no_leave_start_airport_{d}")
    
    # 不可以跑回去睡機場 (抵達)
    # h^a_{n_md} = 0, forall d in D \setminus \{n_d\}
    for d in D:
        if d != num_days - 1: # num_days 是您之前定義的總天數
            model.addConstr(ha[num_hotels - 1, d] == 0, f"no_arrive_end_airport_{d}")
    
    
    # ... (您後續的其他約束，例如每天只能住一間旅館等)
    
    # 第一天返回旅館時間約束 (equation 8)
    model.addConstr(
        arrival_time_day1 +
        gp.quicksum(stay_duration[i] * x[i, 0] for i in I) +
        gp.quicksum(transport_times[k,p,q] * t[p, q, k, 0] for p in P for q in P if p != q for k in K) <= return_time_hotel[0],
        "return_day1"
    )
    
    # 其他天返回旅館時間約束 (equation 9)
    for d in D[1:]:
        model.addConstr(
            lt[d] + 
            gp.quicksum(stay_duration[i] * x[i, d] for i in I) +
            gp.quicksum(transport_times[k,p,q] * t[p, q, k, d] for p in P for q in P if p != q for k in K) <= return_time_hotel[d],
            f"return_day_{d}"
        )

    # 返回旅館時間約束-2 (new constraint 9-2)
    for d in D:
        for m in M:
            model.addConstr(
                a[num_attractions+m,d] <= return_time_hotel[d] + bigM * (1 - ha[m,d]),
                f"return_to_hotel_{m}_{d}"
            )

    # 同一天內地點間移動的時間約束 (equation 10)
    
    #從hotel 出發 (new equation)
    for d in D:
        for q in P:
            travel_time_expr = gp.quicksum(transport_times[k,num_attractions+m,q] * t[num_attractions+m, q, k, d] for k in K for m in M)
            lhs = lt[d]  + travel_time_expr - a[q, d]
            rhs = bigM * (1 - gp.quicksum(t[num_attractions+m, q, k, d] for k in K for m in M))
            #print(f"p=hotel, q={q}, d={d}, lhs={lhs}, rhs={rhs}") # <--- 在這裡添加 print 語句
            model.addConstr(lhs <= rhs, f"travel_time_hotel_{q}_{d}")
    
    #from tourist spots to other places (updated equation 10)
    for i in I:
        for q in P:
            if i != q:
                for d in D:
                    travel_time_expr = gp.quicksum(transport_times[k,i,q] * t[i, q, k, d] for k in K)
                    lhs = a[i, d] + stay_duration[i] + travel_time_expr - a[q, d]
                    rhs = bigM * (1 - gp.quicksum(t[i, q, k, d] for k in K))
                    #print(f"p={i}, q={q}, d={d}, lhs={lhs}, rhs={rhs}") # <--- 在這裡添加 print 語句
                    model.addConstr(lhs <= rhs, f"travel_time_{i}_{q}_{d}")
    
    
    # 每個景點最多參觀一次 (equation 11)
    for i in I:
        model.addConstr(gp.quicksum(x[i, d] for d in D) <= 1, f"visit_once_{i}")
    
    # 每天必須抵達一個旅館(equation 12)
    for d in D:
        model.addConstr(gp.quicksum(ha[m, d] for m in M) == 1, f"arrive_one_hotel_{d}")
    
    # 如果今晚抵達旅館 m，明天早上必須從同一旅館 m 出發(最後一天除外) (equation 13)
    for m in M:
        for d in D[:-1]:
            model.addConstr(ha[m, d] == hl[m, d + 1], f"hotel_continuity_{m}_{d}")
    
    # 第一天早上必須離開機場 (equation 14)
    model.addConstr(hl[0, 0] == 1, "leave_airport_day1")
    
    # 最後一天晚上必須抵達機場 (equation 15)
    model.addConstr(ha[num_hotels - 1, num_days - 1] == 1, "arrive_airport_last_day")
    
    # 離開景點後的去向 (equation 16) - 修改為逐日約束
    for i in I:
        for d in D: # 新增 d 的外層循環
            model.addConstr(
                gp.quicksum(t[i, j, k, d] for j in I for k in K if i != j) + 
                gp.quicksum(t[i, num_attractions + m, k, d] for m in M for k in K) == x[i, d], 
                f"leave_spot_{i}_{d}" # 修改約束名稱以包含 d
            )
    
    # 前往景點的起點 (equation 17) - 修改為逐日約束
    for i in I:
        for d in D: # 新增 d 的外層循環
            model.addConstr(
                gp.quicksum(t[j, i, k, d] for j in I for k in K if i != j) + # 移除 d 的循環
                gp.quicksum(t[num_attractions + m, i, k, d] for m in M for k in K) == x[i, d], # 移除 d 的循環
                f"arrive_spot_{i}_{d}" # 修改約束名稱以包含 d
            )
    
    # 早上離開旅館後的去向 (equation 18) -update:可以留在旅館裡
    for m in M:
        for d in D:
            model.addConstr(
                gp.quicksum(t[num_attractions + m, q, k, d] for q in P for k in K) == hl[m, d],
                f"leave_hotel_to_spot_{m}_{d}"
            )
    
    # 晚上離開景點(旅館)後要去旅館 (equation 19)
    for m in M:
        for d in D:
            model.addConstr(
                gp.quicksum(t[p, num_attractions + m, k, d] for p in P for k in K) == ha[m, d],
                f"arrive_hotel_from_spot_{m}_{d}"
            )
    
    # 每個地方一天只能到達一次 (new equation)
    for d in D:
        for q in P:
            model.addConstr(
                gp.quicksum(t[p, q, k, d] for p in P for k in K) <= 1,
                f"arrive_maximumn_once_at_each_spot{q}_on_each_day_{d}"
            )
    
    # 每個地方只能用一種交通方式抵達 (new equation)
    for d in D:
        for p in P:
            for q in P:
                model.addConstr(
                    gp.quicksum(t[p, q, k, d] for k in K) <= 1,
                    f"one transportation type{p}_{q}_{d}"
                )
    
    #離開時間要大於使用者設定的時間 (new equation)
    for d in D:
        model.addConstr(
            lt[d] >= departure_time_hotel[d],
            f"leavetime_{d}"
        )
    
    
    # ----- 7. 求解模型 -----
    model.optimize()
    # ----- 8. 輸出結果與錯誤處理 -----
    if model.status == GRB.OPTIMAL:
        print("\nOptimal solution found.")
        print(f"最大總幸福值：{model.objVal}")
        print(f"\n耗費時間：{model.Runtime:.2f} seconds")
        return model.objVal, model.Runtime
    elif model.status == GRB.INFEASIBLE:
        print("Model is infeasible (無可行解)")
        model.computeIIS()
        model.write("infeasible.ilp")
        print("Infeasible constraints written to infeasible.ilp")
        return None, None
    else:
        print(f"模型沒有找到最佳解，狀態碼：{model.status}")
        return None, None
```

```python
# 優先選最近的景點
def naive(
    S_I, S_D, S_P, S_K, S_M,
    I, D, P, K, M,
    num_attractions, num_hotels, num_days,
    happiness_values, transport_costs, hotel_costs, spot_costs, budget,
    departure_time_hotel, arrival_time_day1, stay_duration,
    open_times, close_times, return_time_hotel, transport_times
):
    ''' explanation:
    -(leave at departure time)
    -consider only train
    -choose only the cheapest hotel every day in advance
    -always visit places from nearest to farthest (by travel time)
    -if time is not enough for every unvisited place, go back to hotel
    '''
    import numpy as np
    import pandas as pd
    import time

    start_time = time.time()
    
    #spot name and their indices
    spot_index = pd.Series(index=happiness_values.index,data=[i for i in range(0,len(happiness_values))])
    
    #visited spot or not, #x
    visited = pd.Series(index=happiness_values.index,data=[0 for i in range(0,len(happiness_values))])
    
    #arrival time at spot i on day d, spot only!
    a_places = np.zeros((num_attractions, num_days))
    
    #cheapest hotel everyday, first and last hotel are airports
    hotel_df = pd.DataFrame(hotel_costs)
    hotel_df = hotel_df.iloc[:,1:-1].astype(float)
    cheapest_hotels = hotel_df.idxmin(axis=1).tolist()
    cheapest_hotels[num_days-1] = num_hotels - 1 #最後一天要回機場。cheapest_hotels[0]為第一天入住之旅館，cheapest_hotels[num_days-1]為最後一天要到達的機場
    
    #total cost, happiness
    total_cost = sum(hotel_costs[d][cheapest_hotels[d]] for d in range(0,num_days)) #hotel 預先訂好
    total_happiness = 0

    for d in range(0, num_days):
        if d == 0:
            current_time = arrival_time_day1
            current_place = 0 # airport
        else:
            current_time = departure_time_hotel[d]
            current_place = num_attractions + cheapest_hotels[d-1] #起點為前一天入住的hotel
    
        #檢查總花費
        if total_cost >= budget:
            break
        #預留交通費
        min_trans_cost = 0
        for date in range(d,num_days):
            if d == 0:
                min_trans_cost += transport_costs[1,0,num_attractions+cheapest_hotels[0]] #機場到入住旅館
            else:
                min_trans_cost += transport_costs[1, num_attractions+cheapest_hotels[date-1], num_attractions+cheapest_hotels[date]] #1 is the train index
        if  total_cost + min_trans_cost >= budget:
            break

        # ==== POLICY CHANGE: visit closest unvisited spot ====
        train_idx = 1 # 0: Walk, 1: Train, 2: Car -- assumed by context
        candidate_spots = []
        for spot in happiness_values.index:
            idx = spot_index[spot]
            if visited[spot] == 0:
                travel_time = transport_times[train_idx, current_place, idx]
                candidate_spots.append((spot, idx, travel_time))
        # Sort by travel_time, nearest first
        candidate_spots.sort(key=lambda x: x[2])

        cnt = 0 
        for spot, idx, travel_time in candidate_spots:
            travel_cost = transport_costs[train_idx, current_place, idx]
            back_time = transport_times[train_idx, idx, num_attractions + cheapest_hotels[d]]
            back_cost = transport_costs[train_idx, idx, num_attractions + cheapest_hotels[d]]

            spot_open = open_times.iloc[idx, d] if hasattr(open_times, "iloc") else open_times[idx][d]
            spot_close = close_times.iloc[idx, d] if hasattr(close_times, "iloc") else close_times[idx][d]
            spot_stay = stay_duration.iloc[idx] if hasattr(stay_duration, "iloc") else stay_duration[spot]
            spot_happiness = happiness_values.iloc[idx] if hasattr(happiness_values, "iloc") else happiness_values[spot]
            spot_cost = spot_costs.iloc[idx] if hasattr(spot_costs, "iloc") else spot_costs[spot]

            if total_cost + min_trans_cost + travel_cost + spot_cost + back_cost < budget:
                if visited[spot] == 0 and \
                    current_time + travel_time >= spot_open and \
                    current_time + travel_time + spot_stay <= spot_close and \
                    current_time + travel_time + spot_stay + back_time <= return_time_hotel[d]:
                    a_places[idx][d] = current_time + travel_time 
                    visited[spot] = 1
                    current_time = current_time + travel_time + spot_stay
                    total_cost += spot_cost
                    total_cost += travel_cost
                    total_happiness += spot_happiness
                    current_place = idx

            if cnt == num_attractions - 1: ##no proper place to visit, go to hotel 
                # POLICY: only write hotel arrival time for hotel rows
                # a_places only has size (num_attractions, num_days), no hotel rows
                pass
            cnt += 1
    
    # return the objective value and computation time
    computation_time = time.time() - start_time # End the timer
    
    print(f"total happiness = {total_happiness}")
    print(f"total cost = {total_cost}, total budget = {budget}\n")

    return total_happiness, computation_time, total_cost
```

```python
#naive algorithm
def naive_original(
    S_I, S_D, S_P, S_K, S_M,
    I, D, P, K, M,
    num_attractions, num_hotels, num_days,
    happiness_values, transport_costs, hotel_costs, spot_costs, budget,
    departure_time_hotel, arrival_time_day1, stay_duration,
    open_times, close_times, return_time_hotel, transport_times
):
    ''' explanation:
    -(leave at departure time)
    -consider only train
    -choose only the cheapest hotel every day in advance
    -always visit places from high too low happiness
    -if time is not enough for every unvisited place, go back to hotel
    '''
    import numpy as np
    import pandas as pd
    import time

    start_time = time.time()
    
    #spot name and their indices
    spot_index = pd.Series(index=happiness_values.index,data=[i for i in range(0,len(happiness_values))])
    
    #visited spot or not, #x
    visited = pd.Series(index=happiness_values.index,data=[0 for i in range(0,len(happiness_values))])
    
    #arrival time at spot i on day d, spot only!
    a_places = np.zeros((num_places, num_days))
    
    #sort 
    sorted_happiness = happiness_values.sort_values(ascending=False)
    
    #cheapest hotel everyday, first and last hotel are airports
    hotel_df = pd.DataFrame(hotel_costs)
    hotel_df = hotel_df.iloc[:,1:-1].astype(float)
    cheapest_hotels = hotel_df.idxmin(axis=1).tolist()
    cheapest_hotels[num_days-1] = num_hotels - 1 #最後一天要回機場。cheapest_hotels[0]為第一天入住之旅館，cheapest_hotels[num_days-1]為最後一天要到達的機場
    
    
    #total cost, happiness
    total_cost = sum(hotel_costs[d][cheapest_hotels[d]] for d in range(0,num_days)) #hotel 預先訂好
    total_happiness = 0
    
 
    for d in range(0, num_days):
        #print("day",d)
        if d == 0:
            current_time = arrival_time_day1
            current_place = 0 # airport
            #print(f"leave airport at {arrival_time_day1}")
        else:
            current_time = departure_time_hotel[d]
            current_place = num_attractions + cheapest_hotels[d-1] #起點為前一天入住的hotel
            #print(f"leave hotel at {departure_time_hotel[d]}")
    
        #------------------------------------------------ADDED!------------------------------------------------#
        #檢查總花費
        if total_cost >= budget:
            break
        #預留交通費
        min_trans_cost = 0
        for date in range(d,num_days):
            if d == 0:
                min_trans_cost += transport_costs[1,0,num_attractions+cheapest_hotels[0]] #機場到入住旅館
            else:
                min_trans_cost += transport_costs[1, num_attractions+cheapest_hotels[date-1], num_attractions+cheapest_hotels[date]] #1 is the train index
        #print(f"---Day {d}: total cost = {total_cost}, min_trans_cost = {min_trans_cost}---")
        if  total_cost + min_trans_cost >= budget:
            break
        
        #------------------------------------------------END_ADDED!--------------------------------------------#

        #find the nextspot with highst happiness value, which is: 1.unvisited. 2.open time and close time are matched. 3.time after visist is enough to go back to hotel.
        cnt = 0 
        for spot, val in sorted_happiness.items():
            idx = spot_index[spot]
            #print(spot,idx)
            # ==== ADJUSTMENT: transport_times and transport_costs are numpy arrays, use index 1 for Train ====
            train_idx = 1 # 0: Walk, 1: Train, 2: Car -- assumed by context
            travel_time = transport_times[train_idx, current_place, idx]
            travel_cost = transport_costs[train_idx, current_place, idx]
            back_time = transport_times[train_idx, idx, num_attractions + cheapest_hotels[d]]
            back_cost = transport_costs[train_idx, idx, num_attractions + cheapest_hotels[d]]
            # ==== ADJUSTMENT: open_times, close_times, stay_duration, happiness_values, spot_costs are pd.Series or DataFrame ====
            spot_open = open_times.iloc[idx, d] if hasattr(open_times, "iloc") else open_times[idx][d]
            spot_close = close_times.iloc[idx, d] if hasattr(close_times, "iloc") else close_times[idx][d]
            spot_stay = stay_duration.iloc[idx] if hasattr(stay_duration, "iloc") else stay_duration[spot]
            spot_happiness = happiness_values.iloc[idx] if hasattr(happiness_values, "iloc") else happiness_values[spot]
            spot_cost = spot_costs.iloc[idx] if hasattr(spot_costs, "iloc") else spot_costs[spot]
            # ==== ADJUSTMENT END ====

            #檢查錢、再檢查時間
            if total_cost + min_trans_cost + transport_costs[train_idx][current_place][idx] + spot_costs[idx] + transport_costs[train_idx][idx][num_attractions+cheapest_hotels[d]] < budget:
                if visited[spot] == 0 and \
                    current_time + travel_time >= spot_open and \
                    current_time + travel_time + spot_stay <= spot_close and \
                    current_time + travel_time + spot_stay + back_time <= return_time_hotel[d]:
                    a_places[idx][d] = current_time + travel_time 
                    visited[spot] = 1
                    current_time = current_time + travel_time + spot_stay
                    #print(f"on train for {travel_time}")
                    #print(f"visit {spot} at {a_places[idx][d]}")
                    #print(f"stay at {spot} for {spot_stay}")
        
                    total_cost += spot_cost
                    total_cost += travel_cost
                    total_happiness += spot_happiness

                    current_place = idx
                    #print(f"visit spot {idx} with transcost {travel_cost} and spot cost {spot_cost}")

            if cnt == num_attractions - 1: ##no proper place to visit, go to hotel 
                a_places[num_attractions + cheapest_hotels[d]][d] = current_time + back_time
                #print(f"back to hotel {cheapest_hotels[d]} at {a_places[num_attractions + cheapest_hotels[d]][d]} with RH = {return_time_hotel[d]}")
                total_cost += transport_costs[train_idx][current_place][num_attractions + cheapest_hotels[d]]
                #print(f"back to hotel with transcost {transport_costs[train_idx][current_place][num_attractions + cheapest_hotels[d]]}")
            cnt += 1
    
    # return the objective value and computation time
    computation_time = time.time() - start_time # End the timer
    
    print(f"total happiness = {total_happiness}")
    print(f"total cost = {total_cost}, total budget = {budget}\n")

    return total_happiness, computation_time, total_cost
```

### Define Heuristic

```python
#目前最對的
def heuristic(S_I, S_D, S_P, S_K, S_M, I, D, P, K, M,
    num_attractions, num_hotels, num_days,
    happiness_values, transport_costs, hotel_costs, spot_costs, budget,
    departure_time_hotel, arrival_time_day1, stay_duration,
    open_times, close_times, return_time_hotel, transport_times):

    S_I = range(0, len(I))  #  the first hotel is the starting airport and the last hotel is the leaving airport
    S_M = range(len(I), len(P))
    S_P = range(0, len(P))
    S_D = range(0, len(D))
    S_K = range(0, len(K))
    
    import numpy as np
    from sklearn.cluster import DBSCAN
    import time

    start_time = time.time()

    BigM_hotel_cost = 0
    hotel_candidates = list(S_M)[1:-1]  # 去掉機場（第0和最後一個）
    min_hotel_cost = budget
    for d in S_D:
        for h in hotel_candidates:
            cost = hotel_costs[d,h - len(S_I)]  # 飯店索引偏移處理  
            BigM_hotel_cost = max(BigM_hotel_cost, cost)
            if cost < min_hotel_cost:
                day_hotel_id = h
                min_hotel_cost = cost

    BigM_time = 1440
    
    # ==== 使用 P 代替 all_ids ====
    id_to_index = {pid: idx for idx, pid in enumerate(S_P)}
    index_to_id = {idx: pid for pid, idx in id_to_index.items()}

    # ==== 執行 DBSCAN ====
    db = DBSCAN(eps=2, min_samples=1, metric='precomputed')
    labels = db.fit_predict(transport_times[2])  # Use Car as index 2 for numpy array
    
    
    # ==== 建立 cluster_map 和 id_to_cluster ====
    cluster_map = {}
    id_to_cluster = {}
    for idx, label in enumerate(labels):
        pid = S_P[idx]
        cluster_map.setdefault(label, []).append(pid)
        id_to_cluster[pid] = label
    print(cluster_map)

    # ==== 建立 place_score 和 place_score_dict ====
    place_score = []
    for i in S_I:
        h = happiness_values[i] 
        t = stay_duration[i] 
        score = h / (t + 1e-6)  # 避免除以零
        place_score.append((i, score))  # 保留 ID 當 key，後面好查
    place_score.sort(key=lambda x: x[1], reverse=True)
    place_score_dict = dict(place_score)

    # 數字對交通方式
    alpha = 1e-6
    total_cost = min_hotel_cost*(len(D)-1)
    cost_in_day = np.zeros(len(S_D))
    daily_budget = budget / len(S_D)


    ##修改cost_in_day
    # ===== 選擇交通方式函數 =====
    def get_best_travel_by_ratio(p1, p2, d, alpha=1e-6):
        best_score = -float('inf')
        best_time = BigM_time
        best_cost = BigM_hotel_cost
        best_mode = None

        # 計算 walk 的 travel time，作為比較基準
        try:
            t_walk = transport_times[0, p1, p2]  # 0: Walk
        except Exception:
            t_walk = BigM_time

        if t_walk >= BigM_time:
            return best_time, best_cost, best_mode  # 若走路無法到，則整組無效

        for k in S_K:
            t = transport_times[k, p1, p2]
            c = transport_costs[k, p1, p2]
            if t >= BigM_time or c >= BigM_hotel_cost:
                continue
            cost_in_day_after_mode = cost_in_day[d] + c
            time_saving_ratio = (t_walk - t) / (t_walk + alpha)
            cost_ratio = cost_in_day_after_mode / daily_budget
            score = time_saving_ratio / (cost_ratio + alpha)
            if score > best_score:
                best_score = score
                best_time = t
                best_cost = c
                best_mode = k
        return best_time, best_cost, best_mode

    # ===== 幫助函數：取得某 cluster 尚未拜訪景點的總分 =====
    def get_cluster_score(cluster_id, visited_set,d):
        available_cluster_score = 0
        for pid in cluster_map[cluster_id]: 
            if pid in S_I and pid not in visited_set:
                travel_t, travel_c, best_k = get_best_travel_by_ratio(current_location, pid, d)
                visit_t = stay_duration[pid]
                arrival_time = current_time + travel_t
                leave_time = arrival_time + visit_t
                if close_times[pid,d] >= leave_time: 
                    available_cluster_score += place_score_dict.get(pid, 0) 
        return available_cluster_score

    # ===== 幫助函數：從其他 cluster 中找到最近且幸福值最高的 cluster =====
    def get_nearest_high_score_cluster(current_cl, excluded, d):
        candidates = [cl for cl in cluster_map if cl != excluded and cl != current_cl]
        best_score, best_cl = -1, None
        for cl in candidates:
            min_dist = float('inf')
            for p2 in cluster_map[cl]:
                if p2 in visited:
                    continue
                t, _, _ = get_best_travel_by_ratio(current_location, p2, d)
                if t < min_dist:
                    min_dist = t
            score = get_cluster_score(cl,visited,d) / (min_dist + 1e-6)
            if score > best_score:
                best_score, best_cl = score, cl
        return best_cl, get_cluster_score(best_cl,visited,d)

    #在玩完一個景點後決定要不要換cluster 
    def cluster_switching_decision(
        current_cluster,
        visited,
        get_cluster_score,
        get_nearest_high_score_cluster,
        d
    ):
        """
        Decide the next cluster to use based on remaining scores and hotel cluster.
        Returns the new current_cluster to use (may be the same as before).
        """
        remaining_score_in_current = get_cluster_score(current_cluster, visited,d)
        next_best_cluster, next_best_score = get_nearest_high_score_cluster(current_cluster, current_cluster, d)

        if next_best_score > 0 and remaining_score_in_current < next_best_score / 4:
            #print(f"  !! Switching cluster: current {current_cluster} remaining score {remaining_score_in_current:.5f} < 1/4 of cluster {next_best_cluster} score {next_best_score:.5f}")
            return next_best_cluster
        else:
            return current_cluster

    # ==== 計算 cluster 到來機場距離 ====
    airport_start = S_M[0]
    airport_end = S_M[-1]

    # ==== 計算 cluster 幸福分數總和 ====
    cluster_scores = {}
    for label, members in cluster_map.items():
        total_score = sum(place_score_dict.get(pid, 0) for pid in members if pid in S_I)
        cluster_scores[label] = total_score


    # ===== 行程規劃主流程 =====
    itinerary = []
    visited = set()
    total_happiness = 0
    current_cluster = id_to_cluster[airport_start]

    for d in S_D:
        current_time = arrival_time_day1 if d == 0 else departure_time_hotel[d]
        current_location = airport_start if d == 0 else itinerary[-1]['hotel']
        day_plan = []

        new_cluster = cluster_switching_decision(
            current_cluster,
            visited,
            get_cluster_score,
            get_nearest_high_score_cluster,
            d
        )

        if new_cluster != current_cluster:
            #print(f"** Switching to cluster {new_cluster} today, after leaving hotel {current_location}. **")
            current_cluster = new_cluster

        unvisited_in_cluster = [pid for pid in cluster_map[current_cluster] if pid not in visited and pid in S_I]
        #print(unvisited_in_cluster)
        if d == 0:
            if not unvisited_in_cluster:
                earliest_open = arrival_time_day1
            else:
                # 找出這些景點中，當天最早的開門時間
                earliest_open = min(open_times[pid, d] for pid in unvisited_in_cluster)
            current_time = max(arrival_time_day1, earliest_open)
        else:
            # 如果都參觀過，就使用原本 departure_time_hotel[d]
            if not unvisited_in_cluster:
                earliest_open = departure_time_hotel[d]
            else:
                # 找出這些景點中，當天最早的開門時間
                earliest_open = min(open_times[pid, d] for pid in unvisited_in_cluster)
            # current_time 是比較大的那個
            current_time = max(departure_time_hotel[d], earliest_open)

        #print(f"\nDay {d} planning...")
        #print(f"\n==== Day {d} Start ====")
        #print(f"  Start location: {current_location} (可能是機場或前一晚飯店)")
        #print(f"  Start time: {current_time:.2f} min, Budget left: {budget - total_cost:.2f}")
        #print(f"  Visited so far: {visited}")

        # --- Dynamic cluster switching loop ---
        # Use a while loop to allow restarting when cluster switches
        while True:
            cluster_place_list = sorted(
                cluster_map[current_cluster],
                key=lambda x: place_score_dict.get(x, 0),
                reverse=True
            )
            switched = False
            for pid in cluster_place_list:
                if pid in visited or pid not in S_I:
                    continue
                open_t = open_times[pid, d]
                close_t = close_times[pid, d]

                travel_t, travel_c, best_k = get_best_travel_by_ratio(current_location, pid, d)
                visit_t = stay_duration[pid]
                arrival_time = current_time + travel_t
                leave_time = arrival_time + visit_t

  
                if arrival_time < open_t or leave_time > close_t:
                    continue
                cost_at_spot = spot_costs[pid]

                if d == num_days:
                    return_t, _, _ = get_best_travel_by_ratio(pid, airport_end, d)
                else:
                    for h in hotel_candidates:
                        cost = hotel_costs[d,h - len(S_I)]
                        day_hotel_return_t, day_hotel_return_c, _ = get_best_travel_by_ratio(pid, day_hotel_id, d)
                        if cost == min_hotel_cost:
                            h_return_t, h_return_c, _ = get_best_travel_by_ratio(pid, h, d)
                            if day_hotel_return_t > h_return_t:
                                day_hotel_id = h
                                return_t = h_return_t
                                return_c = h_return_c
                            else:
                                return_t = day_hotel_return_t
                                return_c = day_hotel_return_c
                        else:
                            return_t = day_hotel_return_t
                            return_c = day_hotel_return_c

                if current_time + travel_t + visit_t + return_t > return_time_hotel[d]:
                    continue
                if total_cost + travel_c + cost_at_spot + return_c > budget:
                    continue

                # Visit the spot
                visited.add(pid)
                day_plan.append({
                    'place': pid,
                    'arrive': current_time + travel_t,
                    'depart': current_time + travel_t + visit_t,
                    'mode': best_k
                })
                total_cost += travel_c + cost_at_spot
                cost_in_day[d] += travel_c + cost_at_spot
                total_happiness += happiness_values[pid] 
                current_time += travel_t + visit_t
                current_location = pid


                # --- Dynamic cluster switching decision ---
                new_cluster = cluster_switching_decision(
                    current_cluster,
                    visited,
                    get_cluster_score,
                    get_nearest_high_score_cluster,
                    d
                )
                if new_cluster != current_cluster:
                    current_cluster = new_cluster
                    switched = True
                    break  # Restart the while True loop with new cluster
            
            if not switched:
                break  # Done with the day's cluster, exit while loop


        #print(f"==== Day {d} End ====")
        #print(f"  Visited today: {[p['place'] for p in day_plan]}")
        #print(f"  End at hotel: {day_hotel_id}, Total cost so far: {total_cost:.2f}, Total happiness: {total_happiness}")
        #print("-" * 40)

        if d != num_days-1:
            return_travel_time, return_travel_cost, _ = get_best_travel_by_ratio(current_location, day_hotel_id, d)
            current_time += return_travel_time
            total_cost += return_travel_cost  # <-- 添加這行
            itinerary.append({'day': d, 'plan': day_plan, 'hotel': day_hotel_id})
        else: # d == num_days-1 (最後一天)
            return_travel_time, return_travel_cost, _ = get_best_travel_by_ratio(current_location, airport_end, d)
            current_time += return_travel_time
            total_cost += return_travel_cost # <-- 添加這行
            itinerary.append({'day': d, 'plan': day_plan, 'hotel': 'airport'})

    # ===== 輸出總結 =====
    print("\n=== 最終行程 ===")
    for day_info in itinerary:
        print(f"Day {day_info['day']}:")
        for p in day_info['plan']:
            print(f"  Visit {p['place']} from {p['arrive']/60:.2f} to {p['depart']/60:.2f} by mode {p['mode']}")
        print(f"  Hotel: {day_info['hotel']}")
    print(f"\nTotal cost: {total_cost}")
    print(f"Total happiness: {total_happiness}")
    computation_time = time.time() - start_time
    return total_happiness, computation_time, total_cost
```
