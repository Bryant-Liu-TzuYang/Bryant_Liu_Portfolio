```python
import pandas as pd
import os


file_path = "/content/drive/MyDrive/臺大/Data Mining/hw2"
os.chdir(file_path)


df_train = pd.read_csv(f"{file_path}/train_concat.csv")
df_test = pd.read_csv(f"{file_path}/test_concat.csv")
```

```python
X = df_train.drop(['Pid', 'Uid', 'Title', 'Alltags', 'Category', 'Concept', 'Subcategory', 'Postdate', 'img_filepath', 'label', 'text'], axis=1)
y = df_train['label']
print(X.shape)
```

### XGBoost

```python
import xgboost as xgb
from sklearn.metrics import root_mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

dtrain = xgb.DMatrix(X_train, label=y_train, enable_categorical=True)
dtest = xgb.DMatrix(X_test, label=y_test, enable_categorical=True)

params = {
    'objective': 'reg:absoluteerror',
    'device': 'cuda',
    'max_depth': 10,
    'learning_rate': 0.01,
    'colsample_bytree': 0.8,
    'random_state': 42,
}

evals = [(dtrain, 'train'), (dtest, 'validation')]

model = xgb.train(
    params = params,
    dtrain = dtrain,
    num_boost_round = 10000,
    evals = evals,
    verbose_eval = 500,
    early_stopping_rounds = 100,
)

preds = model.predict(dtest)
rmse = root_mean_squared_error(y_test, preds)
mae = mean_absolute_error(y_test, preds)

print(f'RMSE: {rmse:.3f}')
print(f'MAE: {mae:.3f}')
```

```python
# Predict on test set
X_test_final = df_test.drop(['Pid', 'Uid', 'Title', 'Alltags', 'Category', 'Concept', 'Subcategory', 'Postdate', 'img_filepath', 'Predicted', 'text'], axis=1)
print(X_test_final.shape)
dtest = xgb.DMatrix(X_test_final)
y_test_pred = model.predict(dtest)
df_test['Predicted'] = y_test_pred
df_test[['Pid', 'Predicted']].to_csv("submission_xgb.csv", index=False)
print("Predictions saved to submission.csv")
```

```python
from hyperopt import STATUS_OK, Trials, fmin, hp, tpe
import xgboost as xgb
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
import cupy as cp

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

X_train_gpu = cp.array(X_train)
X_test_gpu = cp.array(X_test)
y_train_gpu = cp.array(y_train)
y_test_gpu = cp.array(y_test)

def root_mean_squared_error(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

space = {
    'max_depth': hp.quniform("max_depth", 3, 18, 1),
    # 'gamma': hp.uniform('gamma', 1, 9),
    # 'reg_alpha': hp.quniform('reg_alpha', 40, 180, 1),
    # 'reg_lambda': hp.uniform('reg_lambda', 0, 1),
    'colsample_bytree': hp.uniform('colsample_bytree', 0.5, 1),
    # 'min_child_weight': hp.quniform("min_child_weight", 0, 10, 1),
    'learning_rate': hp.uniform('learning_rate', 0.01, 0.3),
    'n_estimators': 10000,
    'seed': 0,
}


def objective(space):
    model = xgb.XGBRegressor(
        max_depth=int(space['max_depth']),
        # gamma=space['gamma'],
        # reg_alpha=int(space['reg_alpha']),
        # reg_lambda=space['reg_lambda'],
        colsample_bytree=space['colsample_bytree'],
        # min_child_weight=int(space['min_child_weight']),
        n_estimators=10000,
        random_state=0,
        early_stopping_rounds=100,
        eval_metric='mae',
        device='cuda',
        objective='reg:absoluteerror',
        learning_rate=space['learning_rate'],
    )

    evaluation = [(X_train, y_train), (X_test, y_test)]

    model.fit(
        X_train_gpu, y_train_gpu,
        eval_set=[(X_train_gpu, y_train_gpu), (X_test_gpu, y_test_gpu)],
        verbose=False,
    )

    preds = model.predict(X_test)
    mae = mean_absolute_error(y_test, preds)

    print(f'MAE: {mae:.3f}')

    return {'loss': mae, 'status': STATUS_OK}

trials = Trials()

best_hyperparams = fmin(
    fn=objective,
    space=space,
    algo=tpe.suggest,
    max_evals=100,
    trials=trials
)

print("The best hyperparameters are:")
print(best_hyperparams)
```

```python
import xgboost as xgb
from sklearn.metrics import root_mean_squared_error, mean_absolute_error

col = X.columns.tolist()
cols = [str(c) for c in col]

dtrain = xgb.DMatrix(X_train, label=y_train, enable_categorical=True, feature_names=cols)
dtest = xgb.DMatrix(X_test, label=y_test, enable_categorical=True, feature_names=cols)

params = {
    'objective': 'reg:absoluteerror',
    'device': 'cuda',
    'max_depth': 10,
    'learning_rate': 0.01,
    'colsample_bytree': 0.8,
    'random_state': 42,
}

params.update(best_hyperparams)
params['max_depth'] = int(params['max_depth'])

evals = [(dtrain, 'train'), (dtest, 'validation')]

model = xgb.train(
    params = params,
    dtrain = dtrain,
    num_boost_round = 10000,
    evals = evals,
    verbose_eval = 500,
    early_stopping_rounds = 100,
)

preds = model.predict(dtest)
rmse = root_mean_squared_error(y_test, preds)
mae = mean_absolute_error(y_test, preds)\

print(f'RMSE: {rmse:.3f}')
print(f'MAE: {mae:.3f}')
```

```python
# Predict on test set
X_test_final = df_test.drop(['Pid', 'Uid', 'Title', 'Alltags', 'Category', 'Concept', 'Subcategory', 'Postdate', 'img_filepath', 'Predicted', 'text'], axis=1)
dtest = xgb.DMatrix(X_test_final)
y_test_pred = model.predict(dtest)
df_test['Predicted'] = y_test_pred
df_test[['Pid', 'Predicted']].to_csv("submission_tuned.csv", index=False)
print("Predictions saved to submission.csv")
```
