### 將交易、建造日期轉換成屋齡，同時處理奇怪的日期

```python
df_train['deal_year'] = df_train['deal_year'] + 1911
df_test['deal_year'] = df_test['deal_year'] + 1911
df_train['built_year'] = df_train['built_year'] + 1911
df_test['built_year'] = df_test['built_year'] + 1911

ymd = (
    df_train['deal_year'].astype(str) + '-' +
    df_train['deal_month'].astype(str).str.zfill(2) + '-' +
    df_train['deal_day'].astype(str).str.zfill(2)
)
df_train['deal_date'] = pd.to_datetime(ymd, format='%Y-%m-%d', errors='coerce')

ymd = (
    df_test['deal_year'].astype(str) + '-' +
    df_test['deal_month'].astype(str).str.zfill(2) + '-' +
    df_test['deal_day'].astype(str).str.zfill(2)
)
df_test['deal_date'] = pd.to_datetime(ymd, format='%Y-%m-%d', errors='coerce')

ymd = (
    df_train['built_year'].astype(str) + '-' +
    df_train['built_month'].astype(str).str.zfill(2) + '-' +
    df_train['built_day'].astype(str).str.zfill(2)
)
df_train['built_date'] = pd.to_datetime(ymd, format='%Y-%m-%d', errors='coerce')

ymd = (
    df_test['built_year'].astype(str) + '-' +
    df_test['built_month'].astype(str).str.zfill(2) + '-' +
    df_test['built_day'].astype(str).str.zfill(2)
)
df_test['built_date'] = pd.to_datetime(ymd, format='%Y-%m-%d', errors='coerce')

mask_bad = df_train['deal_date'].isna() | df_train['built_date'].isna()
bad_deal_date = df_train[mask_bad]
bad_deal_date
```

```python
# 把所有的日期缺失值都填上中位數
df_train['built_date'] = df_train['built_date'].fillna(df_train['built_date'].median())
df_test['built_date'] = df_test['built_date'].fillna(df_test['built_date'].median())
```

```python
def make_MI_score(X,y):
    from sklearn.feature_selection import mutual_info_regression
    from sklearn.feature_selection import SelectKBest

    # Mutual information
    mi = SelectKBest(mutual_info_regression, k=10)
    mi.fit(X, y)
    return pd.Series(mi.scores_, index=X.columns)

X = df_train.drop(columns=['Price', 'deal_date', 'built_date'])
y = df_train['Price']

mi = make_MI_score(X, y)
mi = mi.sort_values(ascending=False).head(10)
mi = mi.sort_values(ascending=True)
tmp = mi.plot(kind='barh', figsize=(15, 7), fontsize=16)
plt.title('Mutual Information', fontsize=20)
plt.show()
```

```python
# ── imports ────────────────────────────────────────────────────────────────────
from sklearn.ensemble import RandomForestRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error

# ── define models ─────────────────────────────────────────────────────────────
models = {
    "RandomForest": RandomForestRegressor(
        n_estimators=200,
        max_depth=12,
        random_state=42,
        n_jobs=-1
    ),
    "LightGBM": LGBMRegressor(
        n_estimators=500,
        learning_rate=0.01,
        num_leaves=31,
        colsample_bytree=0.8,
        random_state=42,
        n_jobs=-1,
        force_row_wise=True,
        device='gpu',
        gpu_use_dp=True,
        gpu_platform_id=0,
        gpu_device_id=0,
        verbose=-1
    ),
    "CatBoost": CatBoostRegressor(
        iterations=500,
        learning_rate=0.05,
        depth=6,
        eval_metric="MAE",
        random_seed=42,
        task_type="GPU",
        devices='0',      # change if you want multiple GPUs: '0:1', etc.
        verbose=False
    ),
    "SVR": SVR(
        C=1.0,
        epsilon=0.1,
        kernel="rbf"
    ),
}

# ── train & evaluate ──────────────────────────────────────────────────────────
results = []
for name, mdl in models.items():
    print(f"Training {name}...")
    mdl.fit(X_train, y_train)
    preds = mdl.predict(X_test)
    mae  = mean_absolute_error(y_test, preds)
    results.append((name, mae))

# ── summary ───────────────────────────────────────────────────────────────────
import pandas as pd

df_results = pd.DataFrame(results, columns=["Model", "MAE"]).set_index("Model")
print("\nHold-out MAE comparison:")
print(df_results.sort_values("MAE"))

# ── next steps ────────────────────────────────────────────────────────────────
# 1. If LightGBM or CatBoost looks strong, you can run a short RandomizedSearchCV
#    over their learning_rate / num_leaves (LGBM) or depth / l2_leaf_reg (CatBoost).
# 2. Consider a simple stacking ensemble:
#      from sklearn.ensemble import StackingRegressor
#      stack = StackingRegressor(
#          estimators=[
#            ("lgbm", models["LightGBM"]),
#            ("cat",  models["CatBoost"])
#          ],
#          final_estimator=RandomForestRegressor(n_estimators=100, random_state=42),
#          cv=5,
#          n_jobs=-1
#      )
#    stack.fit(X_train, y_train)
#    print("Stacked MAE:", mean_absolute_error(y_test, stack.predict(X_test)))
```

```python
from tqdm import tqdm
from time import perf_counter
import numpy as np
from sklearn.model_selection import ParameterSampler, cross_val_score
from sklearn.pipeline import Pipeline
from lightgbm import LGBMRegressor

# 1) Define your parameter grid as before
param_dist = {
    'lgb__num_leaves': [31, 50, 80, 120],
    'lgb__max_depth': [5, 10, 20, 30],
    'lgb__learning_rate': [0.005, 0.01, 0.05, 0.1],
    'lgb__min_child_samples': [5, 20, 50, 100],
    'lgb__feature_fraction': [0.6, 0.8, 1.0],
    'lgb__bagging_fraction': [0.6, 0.8, 1.0],
}

# 2) Set up the pipeline with GPU‐enabled LightGBM
pipe = Pipeline([
    ('lgb', LGBMRegressor(
        random_state=42,
        n_estimators=1000,
        # --- GPU acceleration ---
        device='gpu',
        gpu_platform_id=0,
        gpu_device_id=0
    ))
])

# 3) Sample n_iter combinations
n_iter = 50
sampler = list(ParameterSampler(param_dist, n_iter=n_iter, random_state=42))

best_mae = np.inf
best_params = None

# 4) Global timer
global_start = perf_counter()

# 5) Loop with tqdm
for params in tqdm(sampler, desc="Hyper‐param search", unit="trial"):
    # set pipeline params
    pipe.set_params(**params)
    
    # time this CV
    t0 = perf_counter()
    scores = cross_val_score(
        pipe, X_train, y_train,
        cv=5,
        scoring='neg_mean_absolute_error',
        n_jobs=-1
    )
    elapsed = perf_counter() - t0
    
    mae = -np.mean(scores)
    
    # report this trial
    tqdm.write(f"MAE: {mae:.2f}  |  time: {elapsed:.1f}s  |  params: {params}")
    
    # track best
    if mae < best_mae:
        best_mae = mae
        best_params = params

total_time = perf_counter() - global_start

print("\n=== Search complete ===")
print(f"Best MAE: {best_mae:.2f}")
print(f"Best params: {best_params}")
print(f"Total search time: {total_time:.1f}s")
```

```python
from hyperopt import STATUS_OK, Trials, fmin, hp, tpe
import xgboost as xgb
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error

def root_mean_squared_error(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

space = {
    'max_depth': hp.quniform("max_depth", 3, 18, 1),
    # 'gamma': hp.uniform('gamma', 1, 9),
    # 'reg_alpha': hp.quniform('reg_alpha', 40, 180, 1),
    # 'reg_lambda': hp.uniform('reg_lambda', 0, 1),
    'colsample_bytree': hp.uniform('colsample_bytree', 0.5, 1),
    # 'min_child_weight': hp.quniform("min_child_weight", 0, 10, 1),
    'learning_rate': hp.uniform('learning_rate', 0.01, 0.3),
    'n_estimators': 10000,
    'seed': 0,
}

def objective(space):
    model = xgb.XGBRegressor(
        max_depth=int(space['max_depth']),
        # gamma=space['gamma'],
        # reg_alpha=int(space['reg_alpha']),
        # reg_lambda=space['reg_lambda'],
        colsample_bytree=space['colsample_bytree'],
        # min_child_weight=int(space['min_child_weight']),
        n_estimators=10000,
        random_state=0,
        early_stopping_rounds=100,
        eval_metric='mae',
        device='cuda',
        objective='reg:absoluteerror',
        learning_rate=space['learning_rate'],
    )

    evaluation = [(X_train, y_train), (X_test, y_test)]

    model.fit(
        X_train, y_train, 
        eval_set=evaluation, 
        verbose=False,
    )

    preds = model.predict(X_test)
    mae = mean_absolute_error(y_test, preds)

    print(f'MAE: {mae:.3f}')

    return {'loss': mae, 'status': STATUS_OK}

trials = Trials()

best_hyperparams = fmin(
    fn=objective,
    space=space,
    algo=tpe.suggest,
    max_evals=100,
    trials=trials
)

print("The best hyperparameters are:")
print(best_hyperparams)
```
