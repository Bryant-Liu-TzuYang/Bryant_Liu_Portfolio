```python
file_path = "Instances/Group2/Instance2_0_real.xlsx"
data = import_data.load_file(file_path)
```

```python
#read data
budget = data['BUDGET']
num_days = data['D']
num_hotels = data['M']  # 減去起始和結束機場
num_attractions = data['I']
num_places = data['P']
num_transport_modes = data['K']

S_M = range(1, data['M'] + 1)  #  the first hotel is the starting airport and the last hotel is the leaving airport
S_I = range(data['M'] + 1, data['P'] + 1)
S_P = range(1, data['P'] + 1)
S_D = range(1, data['D'] + 1)
S_K = range(1, data['K'] + 1)

# 第一天抵達機場時間
arrival_time_day1 = data['Day1Arrival'] # 需要從 Excel 讀取

# 每天離開旅館的時間 (不包含第一天和最後一天)
departure_time_hotel = data['LH'] # 需要從 Excel 讀取
departure_time_hotel = departure_time_hotel.reset_index(drop=True)

# 每天返回旅館的最晚時間
return_time_hotel = data['RH']    # 需要從 Excel 讀取
return_time_hotel = return_time_hotel.reset_index(drop=True)

# 每個景點的停留時間 (表格，索引為景點 i)
stay_duration = data['STAY']      # 需要從 Excel 讀取 (DataFrame 或 dict)

# 每個景點的成本 (表格，索引為景點 i)
spot_costs = data['C_ATTRACTION']         # 需要從 Excel 讀取 (DataFrame 或 dict)

# 每個景點的幸福值 (表格，索引為景點 i)
happiness_values = data['HAPPINESS']    # 需要從 Excel 讀取 (DataFrame 或 dict)

# 每個景點每天的開放時間 (表格，索引為景點 i 和天 d)
open_times = data['ATTRACTION_OPEN']          # 需要從 Excel 讀取 (DataFrame 或 dict)

# 每個景點每天的關閉時間 (表格，索引為景點 i 和天 d)
close_times = data['ATTRACTION_CLOSE']         # 需要從 Excel 讀取 (DataFrame 或 dict)

# 旅館每天的住宿成本 (表格，索引為旅館 m 和天 d，不包含起始和結束機場)
hotel_costs = data['C_HOTEL']        # 需要從 Excel 讀取 (DataFrame 或 dict)

# 地點 p 到地點 q 使用交通方式 k 的交通時間 (表格，索引為 p, q, k)
transport_times = data['DURATION']      # 需要從 Excel 讀取 (DataFrame 或 MultiIndex DataFrame 或 dict)

# 地點 p 到地點 q 使用交通方式 k 的交通成本 (表格，索引為 p, q, k)
transport_costs = data['C_TRAVEL']     # 需要從 Excel 讀取 (DataFrame 或 MultiIndex DataFrame 或 dict)


print(f"Budget: {budget}")
print(f"Number of days: {num_days}")
print(f"Number of hotels: {num_hotels}")
print(f"Number of attractions: {num_attractions}")
print(f"Number of places: {num_places}")
print(f"Number of transport modes: {num_transport_modes}")

print("Sets:")
for i in range(len(S_M)):
    print(f"S_M[{i}]: {S_M[i]}")
for i in range(len(S_I)):
    print(f"S_I[{i}]: {S_I[i]}")
for i in range(len(S_P)):
    print(f"S_P[{i}]: {S_P[i]}")
for i in range(len(S_D)):
    print(f"S_D[{i}]: {S_D[i]}")
for i in range(len(S_K)):
    print(f"S_K[{i}]: {S_K[i]}")


print(f"Day 1 Arrival Time: {arrival_time_day1}")
print(f"Departure Time from Hotel: {departure_time_hotel}")
print(f"Return Time to Hotel: {return_time_hotel}")
print(f"Stay Duration: {stay_duration}")
print(f"Spot Costs: {spot_costs}")
print(f"Happiness Values: {happiness_values}")
print(f"Open Times: {open_times}")
print(f"Close Times: {close_times}")
print(f"Hotel Costs: {hotel_costs}")
print(f"Transport Times: {transport_times}")
print(f"Transport Costs: {transport_costs}")
```

```python
#naive algorithm

''' explanation:
-(leave at departure time)
-consider only train
-choose only the cheapest hotel every day in advance
-always visit places from high too low happiness
-if time is not enough for every unvisited place, go back to hotel
-pay all hotels in advance, have a "minimum transportation cost" (sum of moving from hotel to hotel, hotel to airports)
-if money is not enough, visit no other spots
'''
import numpy as np
import pandas as pd
from time import time
start_time = time()  # Start the timer

#spot name and their indices
spot_index = pd.Series(index=happiness_values.index,data=[i for i in range(0,len(happiness_values))])

#visited spot or not, #x
visited = pd.Series(index=happiness_values.index,data=[0 for i in range(0,len(happiness_values))])

#arrival time at spot i on day d, spot only!
a_places = np.zeros((num_places, num_days))

#sort 
sorted_happiness = happiness_values.sort_values(ascending=False)

#cheapest hotel everyday, first and last hotel are airports
hotel_df = pd.DataFrame(hotel_costs)
hotel_df = hotel_df.iloc[:,1:-1].astype(float)
cheapest_hotels = hotel_df.idxmin(axis=1).tolist()
cheapest_hotels[num_days-1] = num_hotels - 1 #最後一天要回機場。cheapest_hotels[0]為第一天入住之旅館，cheapest_hotels[num_days-1]為最後一天要到達的機場


#total cost, happiness
total_cost = sum(hotel_costs[d][cheapest_hotels[d]] for d in range(0,num_days)) #hotel 預先訂好
total_happiness = 0


for d in range(0, num_days):
    print("day",d)
    if d == 0:
        current_time = arrival_time_day1
        current_place = 0 # airport
        print(f"leave airport at {arrival_time_day1}")
    else:
        current_time = departure_time_hotel[d]
        current_place = num_attractions + cheapest_hotels[d-1] #起點為前一天入住的hotel
        print(f"leave hotel at {departure_time_hotel[d]}")

    #------------------------------------------------ADDED!------------------------------------------------#
    #檢查總花費
    if total_cost >= budget:
        break
    #預留交通費
    min_trans_cost = 0
    for date in range(d,num_days):
        if d == 0:
            min_trans_cost += transport_costs[1,0,num_attractions+cheapest_hotels[0]] #機場到入住旅館
        else:
            min_trans_cost += transport_costs[1, num_attractions+cheapest_hotels[date-1], num_attractions+cheapest_hotels[date]] #1 is the train index
    print(f"---Day {d}: total cost = {total_cost}, min_trans_cost = {min_trans_cost}---")
    if  total_cost + min_trans_cost >= budget:
        break
    
    #------------------------------------------------END_ADDED!--------------------------------------------#

    #find the nextspot with highst happiness value, which is: 1.unvisited. 2.open time and close time are matched. 3.time after visist is enough to go back to hotel.
    cnt = 0 
    for spot, val in sorted_happiness.items():
        idx = spot_index[spot]
        print(spot,idx)
        # ==== ADJUSTMENT: transport_times and transport_costs are numpy arrays, use index 1 for Train ====
        train_idx = 1 # 0: Walk, 1: Train, 2: Car -- assumed by context
        travel_time = transport_times[train_idx, current_place, idx]
        travel_cost = transport_costs[train_idx, current_place, idx]
        back_time = transport_times[train_idx, idx, num_attractions + cheapest_hotels[d]]
        back_cost = transport_costs[train_idx, idx, num_attractions + cheapest_hotels[d]]
        # ==== ADJUSTMENT: open_times, close_times, stay_duration, happiness_values, spot_costs are pd.Series or DataFrame ====
        spot_open = open_times.iloc[idx, d] if hasattr(open_times, "iloc") else open_times[idx][d]
        spot_close = close_times.iloc[idx, d] if hasattr(close_times, "iloc") else close_times[idx][d]
        spot_stay = stay_duration.iloc[idx] if hasattr(stay_duration, "iloc") else stay_duration[spot]
        spot_happiness = happiness_values.iloc[idx] if hasattr(happiness_values, "iloc") else happiness_values[spot]
        spot_cost = spot_costs.iloc[idx] if hasattr(spot_costs, "iloc") else spot_costs[spot]
        # ==== ADJUSTMENT END ====

        #檢查錢、再檢查時間
        if total_cost + min_trans_cost + transport_costs[train_idx][current_place][idx] + spot_costs[idx] + transport_costs[train_idx][idx][num_attractions+cheapest_hotels[d]] < budget:
            if visited[spot] == 0 and \
                current_time + travel_time >= spot_open and \
                current_time + travel_time + spot_stay <= spot_close and \
                current_time + travel_time + spot_stay + back_time <= return_time_hotel[d]:
                a_places[idx][d] = current_time + travel_time 
                visited[spot] = 1
                current_time = current_time + travel_time + spot_stay
                print(f"on train for {travel_time}")
                print(f"visit {spot} at {a_places[idx][d]}")
                print(f"stay at {spot} for {spot_stay}")
    
                total_cost += spot_cost
                total_cost += travel_cost
                total_happiness += spot_happiness

                current_place = idx
                print(f"visit spot {idx} with transcost {travel_cost} and spot cost {spot_cost}")

        if cnt == num_attractions - 1: ##no proper place to visit, go to hotel 
            a_places[num_attractions + cheapest_hotels[d]][d] = current_time + back_time
            print(f"back to hotel {cheapest_hotels[d]} at {a_places[num_attractions + cheapest_hotels[d]][d]} with RH = {return_time_hotel[d]}")
            total_cost += transport_costs[train_idx][current_place][num_attractions + cheapest_hotels[d]]
            print(f"back to hotel with transcost {transport_costs[train_idx][current_place][num_attractions + cheapest_hotels[d]]}")
        cnt += 1

# return the objective value and computation time
computation_time = time.time() - start_time # End the timer

print(f"total happiness = {total_happiness}")
print(f"total cost = {total_cost}, total budget = {budget}\n")
```
